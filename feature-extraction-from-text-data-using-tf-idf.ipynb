```python
import re
corpus = []
no_of_words = 0
for i in range(sdf.shape[0]):
    para = sdf['body'][i]
    sentences = re.sub(r'\n{2,}|\n|।|,', "", para)
    words = sentences.split()
    
#     words = remove_stopwords(words,level='hard')
    words = [word for word in words if word not in st_word]
#     words = [word for word in words if not word in stopwords.words('english')]
    words = [BanglaStemmer().stem(word) for word in words]
    no_of_words += len(words) 
    line = ' '.join(words)
    corpus.append(line)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=1000,ngram_range=(1,3))
X = tv.fit_transform(corpus).toarray()
```
